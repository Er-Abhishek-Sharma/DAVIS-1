```python
import matplotlib.pyplot as plt
```

#### Create Liner chart

```python
# x-axis labels (metrics)
x = ["Likes", "Comments", "Shares", "Reach"]

# y-axis values (numeric values)
y = [540, 32, 45, 5200]

# Create line chart
plt.figure(figsize=(7, 5))
plt.plot(x, y, linestyle='dotted', marker='o', color='blue')

plt.title("Post Engagement Metrics on Instagram (2024-04-05)")
plt.xlabel("Metrics")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

#### Create Bar Chart

```python
plt.title("Post Engagement Metrics on Instagram (2024-04-05)")
plt.bar(x,y)
plt.show()
```

#### Create pie chart

```python
plt.figure(figsize=(6,6))
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=140)
plt.title("Post Engagement Metrics on Instagram (2024-04-05)")
plt.show()
```
